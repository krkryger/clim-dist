```python
import json
from tqdm import tqdm
import spacy
from spacy import displacy
import gensim
from climdist.data import load as data_load
```

```python
with open('../pipeline/weather_events.json') as f:
    events = json.load(f)
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format('../data/models/word2vec_251021/w2v_model.txt')
```

```python
eventslist = []
for lst in events.values():
    eventslist += lst

eventslist = [word for word in eventslist if model.has_index_for(word)]
```

```python
def explore_similars(model, eventslist, topn):
    
    for word in eventslist:
        similars = model.most_similar(word, topn=topn)
        print('\n\n\n\n----', word, '----\n\n')
        for entry in similars:
            print(entry[0])
        choice = input('Continue?')
        if choice == 'n':
            break
```

```python
def add_similar_patterns(model, eventslist):
    
    processed_words = []
    patterns = []
    
    to_go = len(eventslist)
    
    for word in eventslist:
        similars = model.most_similar(word, topn=100)
        print('\n\n\n\n----', word, '----\n\n')
        
        for entry in similars:
            if entry[0] not in processed_words:
                print(entry[0])
                
                choice = 0
                while choice not in ['y', 'n', 'x', 'S']:
                    choice = input('Add? ')
                if choice == 'y':
                    patterns.append({'label': 'WEA', 'pattern': entry[0]})
                    processed_words.append(entry[0])
                elif choice == 'n':
                    processed_words.append(entry[0])
                    pass
                elif choice == 'x':
                    to_go -= 1
                    print(f'\n{to_go} more entries to go')
                    break
                elif choice == 'S':
                    break
                
            else:
                pass
                    
                if choice == 'S':
                    break
                    
        if choice == 'S':
            print(f'\n\nProcessed words: {processed_words}')
            break
                
                    
    return patterns
```

```python
patterns = add_similar_patterns(model, eventslist)
```

```python
len(patterns)
```

```python
def clean_patterns(patterns):
    
    for entry in patterns:
        text = entry['pattern']
        for i in range(0, 10):
            if str(i) in text:
                print(text)
                patterns.remove(entry)
            else:
                pass
                
    return patterns
```

```python
cleaned_patterns = clean_patterns(patterns)
```

```python
with open('../pipeline/ner/ruler_patterns_171121.json', 'w', encoding='utf8') as f:
    json.dump(cleaned_patterns, f)
```

```python
nlp = spacy.load('../data/models/spacy_ner_151121/model-best/')
```

```python
# Similarly, it matters if you add the EntityRuler before or after the statistical entity recognizer:
# if it’s added before, the entity recognizer will take the existing entities into account when making predictions
# https://spacy.io/usage/processing-pipelines

nlp.add_pipe('entity_ruler', before='ner')
nlp.disable_pipe('ner')
ruler = nlp.get_pipe('entity_ruler')
ruler.add_patterns(cleaned_patterns)
```

```python
nlp.pipeline
```

```python
df = data_load('sample', readability=True, heading2=False)
```

```python
found_docs = []

for text in tqdm(df.full_text.iloc[:1000]):
    doc = nlp(text)
    wea_ents = 0
    for ent in doc.ents:
        if ent.label_ == 'WEA':
            wea_ents += 1
    if wea_ents > 5:
        found_docs.append(doc)
    
    
```

```python
len(found_docs)
```

```python
displacy_color_code = {'WEA': '#4cafd9',
                  'PER': '#ffb366',
                  'DAT': '#bf80ff',
                  'LOC': '#a88676'}

displacy_options = {'ents': ['WEA', 'PER', 'DAT', 'LOC'], 'colors': displacy_color_code}
```

```python
for doc in found_docs:
    displacy.render(doc, style='ent', jupyter=True, options=displacy_options)
    print('\n\n------------------------------------------------------\n\n')
```
